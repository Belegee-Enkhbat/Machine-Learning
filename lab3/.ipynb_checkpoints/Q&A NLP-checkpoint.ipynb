```python
def normalizeDataSet(app_set):
    ret = pd.DataFrame(columns=["Title", "Text", "Added by", "Date", "Time limit", "Source limit", "Memory limit", "Cluster", "Languages", "Resource", "URL"])
    dfs_to_concat = []

    for index, row in app_set.iterrows():
        title = row.get("Title", "")
        text = row.get("Text", "")
        added_by = row.get("Added by", "")
        date = row.get("Date", "")
        time_limit = row.get("Time limit", "")
        source_limit = row.get("Source limit", "")
        memory_limit = row.get("Memory limit", "")
        cluster = row.get("Cluster", "")
        languages = row.get("Languages", "")
        resource = row.get("Resource", "")

        data = {'Title': title, 'Text': text, 'Added by': added_by, 'Date': date,
                'Time limit': time_limit, 'Source limit': source_limit, 'Memory limit': memory_limit,
                'Cluster': cluster, 'Languages': languages, 'Resource': resource}

        dfs_to_concat.append(pd.DataFrame(data, index=[0]))  

    if dfs_to_concat:  
        ret = pd.concat(dfs_to_concat, ignore_index=True)  

    return ret
```

```python
app_df = normalizeDataSet(df)
```

```python
from sentence_transformers import SentenceTransformer
sen_model = SentenceTransformer('joeddav/xlm-roberta-large-xnli')
```

```python
sentences = []
y = []
z = []

for index, row in app_df.iterrows():
    sentences.append(row['Title'])
    z.append(row['Cluster'])
    y.append(row['Memory limit'])
```

```python
sentence_vectors = sen_model.encode(sentences)
```

```python
it = 0
x_ = []
for i in range(len(sentence_vectors)):
  B = np.array([z[it]])
  # B 1 x 1
  # Concat sentence vector with m2 value
  O = np.append(sentence_vectors[i], B)
  # O 1 x 100 + 1 x 1 = 1 x 101
  x_.append(O)
  it+=1

# x_ 5000 x 101
# y  5000 x 1
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x_, y)
```

```python
def extractSpace(sentence):
    ans = int(sum(z) / len(z))
    if 'м2' in sentence or 'm2' in sentence:
        sentence = sentence.replace('m2', 'м2')
        sentence = sentence.replace(' м2', 'м2')
        tokens = sentence.split(' ')
        for token in tokens:
            if token.endswith('м2') and len(token) > 2:
                ans = int(token[:len(token) - 2])
                return ans
    return ans

def predictEstimate(sentence):
    p_embeddings = sen_model.encode([sentence])
    d = [extractSpace(sentence)]
    p_ = []
    for emb in p_embeddings:
        B = [d[0]]
        p_.append(np.append(emb, B))
    y_prediction = model.predict(p_)
    return y_prediction[0]
```

```python
test_input = "In this problem you have to calculate the determinant of an N x N matrix whose entries are given by m[i][j] = gcd(i,j)k, 1 ≤ i,j ≤ N. Here gcd(i,j) denotes the greatest common divisor of i and j. As the determinant D can grow very large, you have to print D%1000003."
prediction = predictEstimate(test_input)
print(prediction)
```
