```python
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```python
baseurl = 'https://www.spoj.com/problems/classical/sort=0,start=50'
problems_link_list = []
for i in range(1, 80):
    url = baseurl + str(i)
    print(f"Scraping page {i}: {url}")
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Error {response.status_code} accessing {url}")
        continue
    soup = BeautifulSoup(response.text, "html.parser")
    a_list = soup.find_all("td", {"align": "left"})
    for li in a_list:
        a=li.find('a')
        news_url='https://www.spoj.com/'+a['href']
        problems_link_list.append(news_url)
problems_link_list
```

```python
problems_list=set(problems_link_list)
len(problems_list)
```

```python
it = 0
scrappedData = []

for url in problems_link_list:
    it += 1
    print(it)
    if(it>60):
         break
    response = requests.get(url)
    if response.status_code != 200:
        print(response.status_code)
        print('error ', url)
        continue

    soup = BeautifulSoup(response.text, "html.parser")
    print(url)
    
    title = soup.find("h2", {"id": "problem-name"}).text.strip()
    body1 = soup.find("div", {"id": "problem-body"})
    text = body1.find('p').text.strip()
    
    temp = {'URL': url, 'Title': title, 'Text': text}
    
    tbody = soup.find('tbody')
    if tbody:
        rows = tbody.find_all('tr')
        for row in rows:
            cells = row.find_all('td')
            if len(cells) == 2:
                attribute = cells[0].text.strip()
                value = cells[1].text.strip()
                temp[attribute[:-1]] = value 
    scrappedData.append(temp)
    print(scrappedData)
    

print(scrappedData)
```

```python
import csv

keys = [ 'Title', 'Text', 'Added by', 'Date', 'Time limit', 'Source limit', 'Memory limit', 'Cluster', 'Languages', 'Resource','URL']

with open('output.tsv', 'w', newline='') as f_output:
    tsv_output = csv.DictWriter(f_output, fieldnames=keys, delimiter='\t')
    tsv_output.writeheader()
    for data in scrappedData:
        # Remove extra keys from the data dictionary
        filtered_data = {key: data[key] for key in keys if key in data}
        tsv_output.writerow(filtered_data)
    print("File 'output.tsv' has been created.")
print("finished")
```
